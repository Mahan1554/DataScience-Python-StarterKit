````python


#format Specifiers ={value:flags} format a value based on what flags are inserted


# .(number)f    = round to that many decimal places (fixed point)
# :(number)     = allocate that many spaces
# :03           = allocate and zero pad that many spaces
# :<            = left justify
# :>            = right justify
# :٨            = center align
# :+            = use a plus sign to indicate positive value
# :=            = place sign to leftmost position
# :             = insert a space before positive numbers
# :,            = comma separator



p1=3123456.1416
p2=-98745.65
p3=1212.34
print(f"price1 is ${p1:.2f}") # shows 2 decimal digits
print(f"price2 is ${p2:.1f}")
print(f"price3 is ${p3:10}")# uses 10 spaces for representation
print(f"price4 is ${p1:020}") # fills remaining spaces with zero

#left justified
print(f"price1 is ${p1:<10}")
print(f"price1 is ${p2:<10}")
print(f"price1 is ${p3:<10}")

#right justified
print(f"price1 is ${p1:>10}")
print(f"price1 is ${p2:>10}")
print(f"price1 is ${p3:>10}")

#comma seperator
print(f"price1 is ${p1:,}")
print(f"price1 is ${p2:,}")
print(f"price1 is ${p3:,}")

print(f"price1 is ${p1:,.2f}")

````

````python
#Arbitary arguments

# *args = allows you to pass multiple non-key arguments (Tuple)
# **kwargs = allows you to pass multiple keyword arguments (dictionary)
#           * unpacking operator

def add(*args):
    print(type(args))
    total=0
    for arg in args:
        total+=arg
    return total
print(add(1,2))
print(add(1,2,3,4))


def name(*args):
    for arg in args:
        print(arg,end=" ")
name('maneendra','mahan','bhimireddy')

 
def print_address(**kwargs):
    for value in kwargs.values():
        print(value)

print_address(street='killer street',city='not on earth',state='AP',zip=123456)


def a(**kwargs):
    for i in kwargs.values():
        print(i,end=' ')
a(x=1,y=2,z=3)
````

````python
#super function--> super()
#Function used to give access to the methods of a parent class
#Returns a temporary object of a parent class when used

'''
class A:
    pass
class B(A):
    def __init__(self,a,b):
        self.a=a
        self.b=b
class C(A):
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
'''

#the above code can be written using super() function

class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
class B(A):
    def __init__(self,a,b):
        super().__init__(a,b)
    def sumB(self):
        return self.a+self.b
class C(A):
    def __init__(self,a,b,c):
        super().__init__(a,b)
        self.c=c
    def sumC(self):
        return self.a+self.b+self.c

b=B(2,1)
c=C(1,2,3)
print(b.sumB())
print(c.sumC())
````

````python
"""
 the super() function is used to call a method from a parent or
 superclass. It is commonly used in the context of method overriding 
 in inheritance. The super() function returns a temporary object of
 the superclass, which allows you to call its methods.
"""

class Parent:
    def show(self):
        print("Parent class method")

class Child(Parent):
    def show(self):
        super().show()  # Calls the show method of the Parent class
        print("Child class method")

# Creating an instance of the Child class
child_instance = Child()

# Calling the overridden method in the Child class
child_instance.show()

````

````python
#abstract classes in python
'''
Prevents a user from creating an object of that class + 
compels a user to override abstract methods in a child class

abstract class -> a class which contains one or more abstract methods
abstract method -> a method that has a declaration but does not have an implementation

-->Can't instantiate abstract class A with abstract method f


--> every concrete (non-abstract) subclass of an abstract class must implement
    all the abstract methods declared in the abstract class.
'''

from abc import ABC,abstractmethod

class A(ABC):
    @abstractmethod #class A  function f() undergoes abstract method
    def f1(self): # this function cannot be inherited bcoz class A is abstract class
        pass
    def s(self):
        print('this is not an abstract method ')
    @abstractmethod
    def f2(self): #every abstract methods of a base class must be overridden in its child classes
        print('this is  abstract method')
        
class B(A):
    def f1(self):
        print('class B abs1')
    def f2(self): 
        print('class B abs2')
        
class C(A):
    def f1(self):
        print('class C abs1')
    def f2(self): 
        print('class C abs2')
     
   
b=B()
c=C()

b.f1(),b.f2()
print()
c.f1(),c.f2()
print()
b.s(),c.s()
````

````python
# abstract Class and abstract method

"""
In Python, an abstract class is a class that cannot be instantiated on
its own and is meant to be subclassed by other classes. It may contain
abstract methods, regular methods, or a combination of both. Abstract
methods are methods that do not have an implementation in the abstract 
class and must be implemented by any concrete (i.e., non-abstract) subclass.

To create an abstract class in Python, you can use the abc module, which stands
for Abstract Base Classes.
"""


from abc import ABC, abstractmethod

class Animal(ABC):  # Animal is an abstract class
    @abstractmethod
    def make_sound(self):
        pass

    def move(self):
        print("Moving")
# Uncommenting the next line would result in an error, as you cannot create an instance of an abstract class.
# animal = Animal()


"""
In this example, Animal is an abstract class that contains an abstract method
make_sound() and a regular method move(). The @abstractmethod decorator is used
to denote that make_sound() must be implemented by any concrete subclass.
"""

class Dog(Animal):
    def make_sound(self):
        print("Bark")

# Creating an instance of the Dog class
dog = Dog()
dog.make_sound()  # This will print "Bark"
dog.move()        # This will print "Moving"

"""
The Dog class is a concrete subclass of Animal and must provide an
implementation for the abstract method make_sound(). Once the abstract
method is implemented in the subclass, you can create instances of that subclass.

"""
````

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

    def perimeter(self):
        return 2 * 3.14 * self.radius

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side**2

    def perimeter(self):
        return 4 * self.side


"""
In this example, Shape is an abstract class defining the interface for all shapes.
Concrete subclasses like Circle and Square must provide implementations for the abstract
methods area and perimeter. This ensures that any shape adhering to the Shape interface
must have methods to calculate its area and perimeter.
"""
````

````python
#Duck typing

'''Duck typing is a concept in dynamic programming languages
like Python, where the type or class of an object is 
determined by its behavior (methods and properties) 
rather than its explicit type. The term "duck typing" comes
from the saying, "If it looks like a duck, swims like a duck,
and quacks like a duck, then it probably is a duck." 
In other words, Python does not check the type of an object 
explicitly but instead checks if the object can perform
certain operations or has specific attributes.'''

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Parrot:
    def speak(self):
        return "Squawk!"

# Function that takes any object with a "speak" method
def make_speak(animal):
    return animal.speak()

# Create instances of different classes
dog = Dog()
cat = Cat()
parrot = Parrot()

# Call the make_speak function with different objects
print(make_speak(dog))     # Outputs: Woof!
print(make_speak(cat))     # Outputs: Meow!
print(make_speak(parrot))  # Outputs: Squawk!



'''In this example, we have three different classes 
(Dog, Cat, and Parrot) that all have a speak method.
The make_speak function does not care about the specific 
class of the object passed to it; it only checks if the 
object has a speak method. This demonstrates the idea of
duck typing – as long as an object behaves like it has a
speak method, it can be used with the make_speak function.'''
````

````python
a = [1, 2, 3, 4, 5]
try:
    print(a[0])
    print(a[10])  # This will raise an IndexError
    print(a[4])
except IndexError:
    print("Index out of range")


#Reason:
"""
In the provided code, an exception occurs when trying to access a[10] because the index 10 is out of range for the list a. When an exception is raised inside the try block, the remaining code within the try block is skipped, and the control moves to the except block.
In this case, since an exception occurs at print(a[10]), the execution jumps to the except block, which prints "Out of index." As a result, the code does not reach the print(a[4]) statement inside the try block.
If you want to see the output of a[4] even if an exception occurs earlier in the try block, you should place the print(a[4]) statement outside the try block or include it in a separate try-except block. 
"""
````

````python
# Raise statement

def divide_numbers(a, b):
    try:
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero.")
        result = a / b
        return result
    except Exception as e:
        print(f"An error occurred: {e}")
        # You can choose to re-raise the caught exception or raise a different one.
        raise ValueError("Custom error message") from e

# Example usage:
try:
    result = divide_numbers(10, 2)
    print(f"Result: {result}")
except ValueError as ve:
    print(f"Caught a ValueError: {ve}")
except ZeroDivisionError as zde:
    print(f"Caught a ZeroDivisionError: {zde}")
````
