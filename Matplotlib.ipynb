```python
# generate sin and cos graphs

x=np.linspace(0,5*np.pi,1000)
sin=np.sin(x)
cos=np.cos(x)
#print(x)


plt.plot(x,sin)
plt.title('Sin Graph')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show() # shows the graphs # without this the 2 graphs will be same plane

plt.plot(x,cos)
plt.title('Cos Graph')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()



```

```python
plt.plot(x,sin,label='sin(x)')
plt.plot(x,cos,label='cos(x)')
plt.title('Sin and Cos Graph')
plt.xlabel('x')
plt.ylabel('sin(x) and cos(x)')
plt.legend(loc='upper right') # displays the labels
plt.show()
```

```python
plt.plot(x,sin,label='sin(x)',color='red',linestyle='--')
plt.plot(x,cos,'*')
```

# **Different types of Graphs**

```python
# line plot

x=np.linspace(0,5*np.pi,1000)
sin=np.sin(x)
plt.figure(figsize=(15,5))
plt.plot(x,sin)
plt
```

```python
# multiple line graphs

x=np.linspace(0,5*np.pi,1000)
sin=np.sin(x)
cos=np.cos(x)
plt.figure(figsize=(15,5))
plt.plot(x,sin,color='green',linewidth=5,label='sin')
plt.plot(x,cos,'--',color='red',label='cos',linewidth=5)
plt.title('Sin and cos Graphs')
plt.xlabel('x')
plt.ylabel('sin(x) and cos(x)')
plt.legend()
plt.show()
```

```python
# linestyle commands

# '-'	Solid line
# '--'	Dashed line
# '-.'	Dashed/dotted line
# ':'	Dotted line

x=np.linspace(0,10)

plt.figure(figsize=(15,5))
plt.plot(x,x+1,color='red',linewidth=2,label='solid',linestyle='-')
plt.plot(x,x+3,color='blue',linewidth=2,label='dashed',linestyle='--')
plt.plot(x,x+5,color='green',linewidth=2,label='dash-dot',linestyle='-.')
plt.plot(x,x+7,color='pink',linewidth=2,label='dotted',linestyle=':')
plt.legend()
```

```python
# adjusting the axes  limits

x=np.linspace(0,10)

plt.figure(figsize=(5,5))
plt.xlim([2,7]) #
plt.ylim([10,1000]) # comment them and see to see the diff
plt.plot(x,np.exp(x))
```

```python
# Multiple histograms

x1=np.random.randn(100)
x2=np.random.randn(100)+5

plt.hist(x1,alpha=0.7,bins=10,edgecolor='black')
plt.hist(x2,alpha=0.7,bins=10,edgecolor='black')
plt.title('Multiple Histograms')
plt.xlabel('values')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2D histogram

'''
A 2D histogram is a statistical visualization that represents the distribution of two continuous variables using a grid of bins. Each bin's color intensity or height represents the number of data points that fall into that bin.
It is useful for analyzing density distributions of data in two dimensions, similar to a heatmap or a 3D bar chart.
'''

# to visualize the relationship between two continuous variables

x=np.random.randn(1000)
y=np.random.randn(1000)
plt.figure(figsize=(5,4))
plt.hist2d(y,x,bins=20)
plt.title('2D Histogram')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
# pie charts --A pie chart is a circular chart used to show proportions of different categories. Each slice represents a category, and its size is proportional to its value.

labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [35, 25, 25, 15]  # Percentages or absolute values
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Fruit Distribution")
plt.show()


```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # Explode 1st slice

# Create Pie Chart
plt.figure(figsize=(6,6))  # Set figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=180,
        colors=colors, explode=explode, shadow=True)

# Title
plt.title("Programming Language Popularity")

# Show plot
plt.show()
```

```python
# multiple subplots

x=np.linspace(0,10,50)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

y1 = x**2
y2 = x + 10
y3 = np.sin(x)
y4 = np.exp(x)

axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)

axes[0, 0].set_title('x^2')
axes[0, 1].set_title('x+10')
axes[1, 0].set_title('sin(x)')
axes[1, 1].set_title('e^x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 10, 50)
x2 = np.random.randn(1000)
y1 = np.sin(x1)
y2 = np.exp(x1)
y3 = np.tan(x1)

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))

# Line plots
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('sin(x)')

axes[0, 1].plot(x1, y2)
axes[0, 1].set_title('exp(x)')

axes[0, 2].plot(x1, y3)
axes[0, 2].set_title('tan(x)')

# Histograms
axes[1, 0].hist(x2, bins=10)
axes[1, 0].set_title('Histogram (20 bins)')

axes[1, 1].hist(x2, bins=100)
axes[1, 1].set_title('Histogram (100 bins)')

# Pie chart
axes[1, 2].pie(sizes, labels=labels, autopct='%1.1f%%')
axes[1, 2].set_title('Programming Languages Popularity')

# Adjust layout and show plot
plt.tight_layout()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Define 3D axis

# Plot the 3D curve
ax.plot3D(x, y, z, color='blue')  # Use color='red' instead of 'red' as a string argument

# Show the plot
plt.show()
```
